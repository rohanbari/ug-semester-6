## class_1-3.py

```python
# Slicing in Python

weekdays = ["Mon", "Tue", "Wed", "Thu", "Fri"]

print(weekdays[0])
print(weekdays[0:3])
print(weekdays[:3])
print(weekdays[3:])

print(weekdays[-1])
print(weekdays[::2])
print(weekdays[::-1])

print(list(reversed(weekdays)))
```
