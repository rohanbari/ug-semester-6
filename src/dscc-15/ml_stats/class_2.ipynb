### Calculating 10k x 10k matrix multiplication

```python
import numpy as np
import time

# Create two random 10,000 x 10,000 matrices
A = np.random.rand(10000, 10000)
B = np.random.rand(10000, 10000)

# Measure time for multiplication
start_time = time.time()
C = np.dot(A, B)   # or A @ B
end_time = time.time()

print(f"Elapsed time: {end_time - start_time:.2f} seconds")
```

### Tuples

Tuples are used vastly in CSV data.

```python

digits = (0, 1, "two")
temp = digits

print(id(temp) == id(digits))
digits = digits + (3, 4)

print(temp)
print(digits)

if temp is digits:
    print("True")
else:
    print("False")
```
