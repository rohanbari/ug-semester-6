**Q1.** Python program to find the factorial of a number using recursion.

```python
def factorial(num: int) -> int:
    if num == 0 or num == 1:
        return 1

    return num * factorial(num - 1)


def main() -> None:
    numInput = 0
    try:
        numInput = int(input("Input an integer: "))
    except Exception as e:
        raise e

    print(f"Factorial of {numInput} is {factorial(numInput)}.")


if __name__ == "__main__":
    main()
```
