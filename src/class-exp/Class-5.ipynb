**Q1.** Python program to find the factorial of a number using recursion.

```python
def factorial(num: int) -> int:
    if num == 0 or num == 1:
        return 1

    return num * factorial(num - 1)


def main() -> None:
    numInput = 0
    try:
        numInput = int(input("Input an integer: "))
    except Exception as e:
        raise e

    print(f"Factorial of {numInput} is {factorial(numInput)}.")


if __name__ == "__main__":
    main()
```

**Q3.** Python program to merge the content of two files.

```python
FILES = [
    "Class-5-res.txt",
    "Class-5-res-more.txt",
    "Class-5-res-other.txt"
]

FILE_OUT = "Class-5-merged.txt"

with open(FILE_OUT, "w") as outFile:
    for fileName in FILES:
        with open(fileName, "r") as inFile:
            outFile.write(inFile.read())
            outFile.write("\n")

print("Write successful. Reading content:")
with open(FILE_OUT, "r") as out:
    print(out.read())
```
