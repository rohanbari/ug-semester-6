### Theory

Different types of data structures are used in their respective places. For
example, if values need to be distinct, then we use `set()`, when memoization
is needed, then `map()` should be used, or generally `list()`, etc.

```python
sampleList = [1, 3.14, "baba", True]
types = [type(sample) for sample in sampleList]
print(types)

strList = ["Rohan", "Dipak", "Sudhanshu", "Ankeet Beta"]
print(strList)

# Display average of numerical data
nums = [num for num in sampleList if isinstance(num, (int, float))]
average = sum(nums) / len(nums)
print(f"Total sum of numerical values: {average:.2f}")
```

```python
a = [1, -8, 5.6, 80, -6.21, 172.60]

print(a[:3:2])
print(a[::-1])

print(a[6:0:-1])
print(a.index(80))
```
